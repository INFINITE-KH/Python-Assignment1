```python
filename = "./complete.csv"

with open (filename) as file:
    content = file.readlines()
#Check helper variable to discount the first element in the list
check = False
#Empty list to insert all nationalities into
nationalities = []
#Looping all items in our list
for item in content:
    #If check is true then it is NOT the first element, thus we are interested in the item
    if(check):
        #Split the item into a temporary variable
        temp = item.split(",")
        #Add the nationality of the current item into our list of nationalities
        nationalities.append(temp[14])
    else:
        #If check is false, it is our first element, and check is set to true, so all other elements are counted.
        check = True
#Counter creates a sorted list of the frequencies of the string in the list
#meaning we have a list of the frequencies of the nationalities, where the first is the most common, and the last
#is the least common.
frequencies = Counter(nationalities)
print ('Most common:')
#Prints the 5 most common nationalities
for letter, count in frequencies.most_common(5):
    print ('%s: %7d' % (letter, count))
```

```python
filename = "./complete.csv"

with open (filename) as file:
    content = file.readlines()
#Check helper variable to discount the first element in the list
check = False
#Empty list to insert all our entries in
ages = []
weights = []
heights = []
#Looping all items in our list
for item in content:
    #If check is true then it is NOT the first element, thus we are interested in the item
    if(check):
        #Split the item into a temporary variable
        temp = item.split(",")
        #Add the age, height and weight of the current item into our lists
        ages.append(temp[6])
        heights.append(temp[9])
        weights.append(temp[10])
    else:
        #If check is false, it is our first element, and check is set to true, so all other elements are counted.
        check = True
#Counter creates a sorted list of the frequencies of the string in the list
#meaning we have a list of the frequencies of the 3 values, where the first is the most common, and the last
#is the least common.
agef = Counter(ages)
heightsf = Counter(heights)
weightsf = Counter(weights)
print ('Most common ages:')
#Prints the most common ages
for letter, count in agef.most_common():
    print ('%s: %7d' % (letter, count))
print ('Most common weights:')
#Prints the most common weights
for letter, count in weightsf.most_common():
    print ('%s: %7d' % (letter, count))
print ('Most common heights:')
#Prints the most common heights
for letter, count in heightsf.most_common():
    print ('%s: %7d' % (letter, count))
```
